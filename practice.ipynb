```python
#quiz
questions={
    'what is the city that you live in?':'birmingham',
    'which month will ramadan start?':'february',
    'which month is eid?':'march',
    'when is christmas?': 'december'
} #this is a dictionary

score=0 #this is a tracker, no questions have been answered yet

for question in questions:
    answer=input(question) #stores users answers in a variable
    
    if answer.lower()==questions[question]:
        score +=1 #updates score
    else:
        print('wrong')

print(f"Quiz finished! Your score: {score}/{len(questions)}")
```

```python
import random

questionss={
    'what month are we in':'feb',
    'what month follows':'march',
    'what follows that month':'april'
}
score=0

questionss_list=list(questionss.keys()) # Convert questions to a list 
random.shuffle(questionss_list) #shuffles the questions

for questions in questionss_list:
    solution=input(questions)
    if solution.lower()==questionss[questions]:
        score += 1
    else:
        print('wrong')


print(f'finished! your score is {score}/{len(questionss)}')
```
